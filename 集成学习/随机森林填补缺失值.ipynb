```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
# from sklearn.model_selection import cross_val_score
dataset = load_boston()
```

```python
x_full,y_full = dataset.data,dataset.target
n_samples = x_full.shape[0]
n_features = x_full.shape[1]
```

```python
#下面将原始数据集做成缺失的数据集
#首先要确定希望的放入缺失数据的比例，假设是50%，总共就是3289个数据缺失
rng = np.random.RandomState(0)
missing_rate = 0.5
missing_samples = int(np.floor(n_samples*n_features*missing_rate))
#floor 向下取整，避免缺失个数出现了小数
```

```python
missing_samples
```

```python
missing_y = rng.randint(0,n_features,missing_samples)
missing_x = rng.randint(0,n_samples,missing_samples)

x_missing = x_full.copy()
y_missing = y_full.copy()
x_missing[missing_x,missing_y] = np.nan
x_missing = pd.DataFrame(x_missing)
```

```python
x_missing.info
```

```python
#使用均值填补
imp_mean = SimpleImputer(missing_values=np.nan,strategy='mean')
x_missing_mean = imp_mean.fit_transform(x_missing)
```

```python
#使用0进行填补
imp_0 = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0)
x_missing_0 = imp_0.fit_transform(x_missing)
```

```python
pd.DataFrame(x_missing_mean).isnull().sum()
```

```python
x_missing
```

```python
#使用随机森林填补
#原理：用没有缺失值的样本做训练集，预测缺失值。
#若有多个缺失的特征列，则用缺失最少的一列先预（因为预测缺失最少的所需要的信息最少）
#填补一个缺失时，先将其他的缺失值填补成0
#遍历所有的特征之后，数据就完整了
x_missing_reg = x_missing.copy()
#找出数据集中，缺失值个数从小到大排列的顺序
#转化为布尔型之后按列加和，也就是每一列的缺失值数量，之后排序取值
#np.sort和argsort的区别是，一个是对值排序，一个是对相应的索引排序。
sortindex = np.argsort(x_missing_reg.isnull().sum(axis=0)).values
for i in sortindex:
    
    #构建我们的新特征矩阵和新标签
    df = x_missing_reg
    fillc = df.iloc[:,i] #取出对应的要填补的列
    #df.columns != i
    df = pd.concat([df[df != fillc],pd.DataFrame(y_full)])
    
    #将含有缺失值的列进行0的填补
    df_0 = SimpleImputer(missing_values=np.nan,
                        strategy='constant',fill_value=0).fit_transform(df)
    
    #找出labels
    ytrain = fillc[fillc.notnull()]
    ytest = fillc[fillc.isnull()]
    #找出训练data，划分训练集测试集
    xtrain = df_0[ytrain.index,:]
    xtest = df_0[ytest.index,:]
    rfc = RandomForestRegressor(n_estimators=100)
    rfc = rfc.fit(xtrain,ytrain)
    ypredict = rfc.predict(xtest)
    x_missing_reg.loc[x_missing_reg.iloc[:,i].isnull(),i]=ypredict
    
```
