```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
data = pd.read_csv(r'F:\My_project\mine_happiness\sklearn课堂\data\titanic_train.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
#数据预处理与特征方程
  #删除cabin,name,ticket列
data.drop(['Cabin','Ticket','PassengerId','Name'],inplace=True,axis=1)
data
```

```python
#填补缺失值
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.head()
```

```python
labels = data["Embarked"].unique().tolist()
data["Embarked"] = data["Embarked"].apply(lambda x: labels.index(x))
```

```python
data["Sex"]=="male"
```

```python
(data["Sex"]=="male").astype('int')
```

```python
data["Sex"] = (data["Sex"]=="male").astype('int')
```

```python
data.loc[:,'Sex'],data.iloc[:,3]
```

```python
x = data.loc[:,data.columns!="Survived"]
y = data.loc[:,data.columns=="Survived"]
```

```python
x.shape,y.shape
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size=0.3)
```

```python
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.index = range(i.shape[0])
Xtest
```

```python
#普通测试
clf = DecisionTreeClassifier(random_state=25)
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
#交叉验证
clf = DecisionTreeClassifier(random_state=25)
score = cross_val_score(clf,x,y,cv=10).mean()
score
```

```python
#学习曲线
tr = []
te = []
for i in range(10):
    clf = DecisionTreeClassifier(random_state=25
                                 ,max_depth=i+1
                                 ,criterion='entropy'
                                )
    clf = clf.fit(Xtrain,Ytrain)
    score_tr = clf.score(Xtrain,Ytrain)
    score_te = clf.score(Xtest,Ytest)
    tr.append(score_tr)
    te.append(score_te)
plt.figure()
plt.plot(range(1,11),tr,color="red",label="train")
plt.plot(range(1,11),te,color="blue",label="test")
plt.xticks(range(0,11))
print('max_train:',max(tr),'\nmax_test:',max(te))
plt.legend()
plt.show()
```
